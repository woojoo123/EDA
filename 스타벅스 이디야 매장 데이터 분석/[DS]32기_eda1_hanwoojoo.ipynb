```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
url = 'https://www.starbucks.co.kr/store/store_map.do?disp=locale'
driver = webdriver.Chrome()
driver.get(url)
```

```python
# 서울 지역 선택
seoul_select = driver.find_element(By.CSS_SELECTOR, '#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1) > a')
seoul_select.click()
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

stores = soup.select('.quickSearchResultBoxSidoGugun > li')

for store in stores:
    print(store.get_text())
```

- BeautifulSoup으로 가져온 매장 리스트에서 매장명과 주소를 각각 추출.
- 매장명은 strong 태그에서, 주소는 p class="result_details" 태그에서 추출.
- 추출한 데이터는 리스트(store_names와 addresses)에 저장.

```python
import pandas as pd

store_names = []  
addresses = []    

# 매장 정보 추출
for store in stores:
    name_tag = store.find('strong') 
    if name_tag:
        store_names.append(name_tag.get_text(strip=True))  # 공백 제거 후 저장
    else:
        store_names.append("매장명 없음")
    
    address_tag = store.find('p', class_='result_details')  
    if address_tag:
        addresses.append(address_tag.get_text(strip=True))  
    else:
        addresses.append("주소 없음")
```

```python
seoul_starbucks_df = pd.DataFrame({
    '매장명': store_names,
    '주소': addresses
})
```

```python
seoul_starbucks_df['구'] = seoul_starbucks_df['주소'].str.extract(r'서울특별시\s(.*?구)')
```

```python
# 웹 페이지 접속

url = 'https://www.ediya.com/contents/find_store.html'
driver = webdriver.Chrome()
driver.get(url)
```

```python
# '주소' 버튼 클릭

find_address = driver.find_element(By.CSS_SELECTOR, '#contentWrap > div.contents > div > div.store_search_pop > ul > li:nth-child(2) > a')
find_address.click()
```

```python
# 서울시 구 리스트 가져오기

gu_list = seoul_starbucks_df['구'].unique()
gu_list
```

```python
ediya_list = []

for gu in gu_list:
    # 검색창 초기화 및 구 이름 입력
    keyword = driver.find_element(By.ID, "keyword")
    keyword.clear()
    keyword.send_keys(f'서울 {gu}')
    
    # 검색 버튼 클릭
    search_button = driver.find_element(By.CSS_SELECTOR, '#keyword_div > form > button')
    search_button.click()
    time.sleep(2)  
    
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    
    # 매장 리스트 추출
    ul_tag = soup.find("ul", id="placesList")
    if not ul_tag:  # 검색 결과가 없는 경우
        print(f"{gu}에 매장이 없습니다.")
        continue

    store_list = ul_tag.find_all("dl")
    
    # 각 매장 정보 추출
    for store in store_list:
        name = store.find("dt").get_text(strip=True)  # 매장명
        address = store.find("dd").get_text(strip=True)  # 주소
        
        # 구 정보 추출
        address_parts = address.split(" ")
        store_gu = address_parts[1] if len(address_parts) >= 2 else "구 정보 없음"
        
        # 데이터 저장
        ediya_list.append({
            "매장명": name,
            "주소": address,
            "구": store_gu
        })

# DataFrame 생성
seoul_ediya_df = pd.DataFrame(ediya_list)

# 데이터 확인
print(seoul_ediya_df)
```

```python
seoul_ediya_df.to_csv("ediya_coffee_seoul.csv", index=False)
```

## 위도 및 경도 변환
- Google Maps API를 사용하여 매장 주소를 위도와 경도로 변환.
- 이를 통해 매장 간의 물리적 거리를 계산 가능.

```python
# pip install googlemaps
```

```python
import pandas as pd
import googlemaps
from geopy.distance import geodesic
import folium

gmaps = googlemaps.Client(key="AIzaSyCIurBd9l7qbK3SwkuRxyt7RT8o55iR1Ew")

ediya_df = pd.read_csv("C:\my_ws\과제\스타벅스 이디야 매장 데이터 분석\ediya_coffee_seoul.csv")  # 이디야 매장 데이터
starbucks_df = pd.read_csv("C:\my_ws\과제\스타벅스 이디야 매장 데이터 분석\starbucks_coffee_seoul.csv")  # 스타벅스 매장 데이터

# Google Maps API를 사용해 위도/경도 변환
def get_lat_lon_google(address):
    try:
        geocode_result = gmaps.geocode(address)
        if geocode_result:
            location = geocode_result[0]['geometry']['location']
            return (location['lat'], location['lng'])
        else:
            return (None, None)
    except Exception as e:
        print(f"Error with address '{address}': {e}")
        return (None, None)

# 이디야 매장 위도/경도 추가
ediya_df[['위도', '경도']] = ediya_df['주소'].apply(lambda x: pd.Series(get_lat_lon_google(x)))

# 스타벅스 매장 위도/경도 추가
starbucks_df[['위도', '경도']] = starbucks_df['주소'].apply(lambda x: pd.Series(get_lat_lon_google(x)))

# 데이터 확인
print(ediya_df.head())
print(starbucks_df.head())
```

## 거리 계산
- 이디야 매장에서 가장 가까운 스타벅스 매장까지의 거리를 계산.
- `geopy` 라이브러리의 `geodesic()` 함수를 사용하여 두 좌표 간의 거리(미터 단위)를 계산.

```python
from geopy.distance import geodesic

# 거리 계산 함수
def calculate_distance(lat1, lon1, lat2, lon2):
    if None in [lat1, lon1, lat2, lon2]:
        return None
    return geodesic((lat1, lon1), (lat2, lon2)).meters

# 각 이디야 매장별로 가장 가까운 스타벅스 거리 계산
closest_distances = []
for _, ediya_row in ediya_df.iterrows():
    ediya_location = (ediya_row['위도'], ediya_row['경도'])
    distances = []
    for _, starbucks_row in starbucks_df.iterrows():
        starbucks_location = (starbucks_row['위도'], starbucks_row['경도'])
        distance = calculate_distance(*ediya_location, *starbucks_location)
        distances.append(distance)
    closest_distances.append(min(distances) if distances else None)

# 결과 추가
ediya_df['가장 가까운 스타벅스 거리(m)'] = closest_distances
ediya_df['500m 이내 여부'] = ediya_df['가장 가까운 스타벅스 거리(m)'] <= 500

# 확인
print(ediya_df[['매장명', '주소', '가장 가까운 스타벅스 거리(m)', '500m 이내 여부']])
```

## 시각화
- 이디야 : 파란색, 스타벅스 : 초록색
- Folium을 사용해 이디야와 스타벅스 매장을 서울 지도에 표시.
- 시각적으로 매장 간의 위치 관계를 확인.

```python
# 서울 중심 좌표
seoul_map = folium.Map(location=[37.5665, 126.9780], zoom_start=12)

# 이디야 매장 추가
for _, row in ediya_df.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=f"이디야: {row['매장명']} ({row['가장 가까운 스타벅스 거리(m)']}m)",
        icon=folium.Icon(color="blue", icon="coffee"),
    ).add_to(seoul_map)

# 스타벅스 매장 추가
for _, row in starbucks_df.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=f"스타벅스: {row['매장명']}",
        icon=folium.Icon(color="green", icon="coffee"),
    ).add_to(seoul_map)

# 지도 저장
seoul_map.save("seoul_coffee_map.html")
print("지도 저장 완료: seoul_coffee_map.html")
```

## 분석 기준

- 고객은 일반적으로 도보로 약 **5~10분(약 500m)** 이내의 거리를 "가까움"으로 인식합니다.
- **국내 상권 분석**에서는 **500m 반경**을 주요 **1차 상권**으로 정의하며, 매장 간의 유동 인구를 분석하는 데 활용합니다.  
  - 출처: 소상공인시장진흥공단 상권정보시스템, 서울시 상권분석 서비스
- **유통 연구**에 따르면, 고객이 가장 선호하는 매장 간 거리는 **300~500m**입니다.  
  - 출처: Clarke (2000), *Retail Location and Consumer Behavior*

따라서, **500m 이내**는 매장 간 "의도적 가까움"을 판단하기에 적절한 기준입니다.

- 500m 이내의 근접성 : 이디야 매장과 가장 가까운 스타벅스 매장이 **500m 이내**에 있는지를 기준으로 근접성을 분석
- 평균 거리 : 매장 간 평균 거리를 계산하여 전반적인 위치 관계를 파악
- 500m 이상 떨어진 매장 : 독립적인 위치 전략을 가진 매장의 비율을 분석

```python
# 500m 이내 비율
within_500m_ratio = ediya_df['500m 이내 여부'].mean() * 100
print(f"이디야 매장의 {within_500m_ratio:.2f}%가 스타벅스 매장 500m 이내에 위치합니다.")

# 거리 통계
print(ediya_df['가장 가까운 스타벅스 거리(m)'].describe())

# 500m 이상 떨어진 매장 수
far_from_starbucks = ediya_df[ediya_df['500m 이내 여부'] == False]
print(f"500m 이상 떨어진 이디야 매장 수: {len(far_from_starbucks)}")
```

# 분석 결과

- 500m 이내 비율 : 전체 이디야 매장 중 **68.65**%가 스타벅스 매장 **500m 이내**에 위치


- 평균 거리 : 이디야 매장과 가장 가까운 스타벅스 매장 간 **평균 거리는 약 408m**


- 500m 이상 거리 : 약 **31.35**%의 이디야 매장은 스타벅스 매장과 **500m 이상** 떨어져 있음
